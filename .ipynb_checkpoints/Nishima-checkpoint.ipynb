# 2 Plot 'Argentina-CABA' and 'Argentina-Catamarca' for number of cases values for entire duration to identify the difference.

```python
df1=data[(data.location=="Argentina-CABA")|(data.location=="Argentina-Catamarca")]
```

```python
df1=df1.fillna(0)
```

```python
import numpy as np
```

```python
df1=df1[df1.unit=="cases"]
```

```python
df1=df1.groupby("location")
```

```python
x=df1.get_group("Argentina-CABA")
x=list(map(int ,x['value']))
x=sum(x)
y=df1.get_group("Argentina-Catamarca")
y=list(map(int ,y['value']))
y=sum(y)
import matplotlib.pyplot as plt
l=[x,y]
s=["Argentina-CABA","Argentina-Catamarca"]
plt.bar(s,l)
plt.title("Difference b/w two")
plt.ylabel("Sum of values")
plt.xlabel("Location")
plt.show()
```

### 3. Plot how data field values are distributed across the locations 'Argentina-Buenos_Aires','Argentina-CABA', 'Argentina-Catamarca'

```python
df1=data[(data.location=="Argentina-CABA")|(data.location=="Argentina-Catamarca")|(data.location=="Argentina-Buenos_Aires")]
group=df1.groupby("location")
k=["Argentina-CABA","Argentina-Catamarca","Argentina-Buenos_Aires"]
for a in k:
    g=group.get_group(a)
    
    h=g["data_field"].value_counts()
    h=list(h.index)
    g1=g.groupby("data_field")
    l=[]
    s=[]
    colors=["grey",'magenta','orange',"r","orange","b"]
    for i in h:
        t=g1.get_group(i)
        x=list(map(int ,t['value']))
        l.append(sum(x))
    print(a)
    plt.figure(figsize=(5,5))
    plt.pie(l,colors=colors,shadow=True, autopct="%0.0f%%", startangle=90,textprops={'fontsize': 9},labeldistance=1.1)
    plt.legend(loc=(1,1),labels=h)
    plt.show()
```

### 4.Build a program to predict the correct word for wrongly typed word. for example whst => what, numbr =>number, good => good

```python
import re
from collections import Counter
import difflib
word=input("Enter a word     : ")
with open("big.txt",'r')as f: 
    text=f.read()
    all_words=re.findall(r'\w+', text.lower())
    list_words=Counter(all_words)
correct_word=difflib.get_close_matches(word, list_words,6)
def Probability(correct_word, N=sum(list_words.values())): 
    return list_words[correct_word] / N
def correction(word): 
    return max(word, key=Probability)
corrected=correction(correct_word)
print("correct words is : "+corrected)
```
